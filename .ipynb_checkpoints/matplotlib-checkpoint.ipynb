```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
#print(x)
ax.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x));
```

```python
plt.plot(x,np.sin(x));
plt.plot(x,np.cos(x));
```

```python
plt.plot(x,np.sin(x-0),color='blue')            #specify by color name
plt.plot(x,np.sin(x-1),color='g')               #shortcolor code(rgbcmyk)
plt.plot(x,np.sin(x-2),color='0.75')            #graysacle b/w 0 & 1
plt.plot(x,np.sin(x-3),color='#FFDD44')         #hex code (RRGGBB fromm 00 to FF)
plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3));    #RGB tuple, values 0 to 1
```

```python
plt.plot(x,np.sin(x-0),color='blue',linestyle='dotted')            #dotted
plt.plot(x,np.sin(x-1),color='g',linestyle='dashed')               #dashesd
plt.plot(x,np.sin(x-2),color='0.75',linestyle='dashdot')           #dashdot
plt.plot(x,np.sin(x-3),color='#FFDD44',linestyle=':')              #dotted
plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3),linestyle='solid');     #solid
```

```python
plt.plot(x,np.sin(x-0),'-g')            #solid green
plt.plot(x,np.sin(x-1),'--c')           #dash cyan
plt.plot(x,np.sin(x-2),'-.k')           #dash dot black
plt.plot(x,np.sin(x-3),':r');           #dotted red
```

```python
plt.plot(x,np.sin(x-0),'-c')
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
#OR TOGETHER
plt.plot(x,np.sin(x-0),'-c')

plt.axis([-1,11,-1.5,1.5]) #min[xmin,xmax,ymin,ymax]

#axes is the fig or entire thing
#axis is x-axis & y-axis
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)

plt.plot(x,y,'*',color='r') #put 0,p,*,+,- etc... for symbol
```

```python
plt.plot(x,y,'ok-'); #put 0,p,*,+,- etc... for symbol  and k in for for color - for line type
#;  to avoid [<matplotlib.lines.Line2D at 0x27d4b950048>] this type of address
```

```python
plt.plot(x,y,'-p',color='gray',
        markersize=12,linewidth=3,   #size of pentagon #width of line
        markerfacecolor='pink', #inside marker
        markeredgecolor='blue',       #edge color of marker
        markeredgewidth=2)

plt.ylim(-1.2,1.2)
```

```python
#SCATTER PLOT
#I can customize individual points
#but in
#normal plot general chnaging
#so normal plots for lot of points
plt.scatter(x,y,marker='o')
```

```python
plt.scatter(x,y,marker='p')
```

```python
x1=np.random.randn(100)
y1=np.random.randn(100)

colors=np.random.randn(100)
sizes=1000*np.random.randn(100)

plt.scatter(x1,y1,c=colors,s=sizes,alpha=0.3,cmap='viridis')
#alpha transperency#cmap virtualization 
plt.colorbar(); #show color scale
```

```python
x1=np.random.randn(100)
y1=np.random.randn(100)

colors=np.random.randn(100)
sizes=1000*np.random.randn(100)

plt.scatter(x1,y1,c=colors,s=sizes,alpha=0.3,cmap='viridis_r') #_r to reverse color bar
#alpha transperency#cmap virtualization 
plt.colorbar(); #show color scale
```

```python
x1=np.random.randn(100)
y1=np.random.randn(100)

colors=np.random.randn(100)
sizes=1000*np.random.randn(100)

plt.scatter(x1,y1,c=colors,s=sizes,alpha=0.3,cmap='flag') 
#alpha transperency#cmap virtualization 
#cmap types
#perceptually uniform sequential
#['inferno' ,'magma','virdis']
plt.colorbar(); #show color scale
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
features=iris.data.T
iris.feature_names
```

```python
plt.scatter(features[0],features[1],alpha=0.2,
           s=100*features[3],c=iris.target,cmap='viridis')
#??iris.target??
plt.title('Sample Plot')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
x1=np.random.normal(0,0.8,1000)
x2=np.random.normal(-2,0.4,1000)
x3=np.random.normal(2,1.6,1000)

kwargs=dict(histtype='stepfilled',alpha=0.5,normed=True,bins=10)#keyword argument #so no need to type again and again

plt.hist(x1,**kwargs)
plt.hist(x2,**kwargs)
plt.hist(x3,**kwargs);
```

```python
plt.hist(x1,**kwargs);
```

```python
#2D HISTOGRAM
x=np.random.normal(2,1,100)
y=np.random.normal(1,1,100)

plt.hist2d(x,y,bins=30,cmap='Blues');
cb=plt.colorbar()
cb.set_label("counts in bin")
```

```python
plt.hist(x1,bins=30)
plt.hist(x2,bins=30);
plt.legend(["first","second"]);
```

```python
plt.hist(x1,bins=30)
plt.hist(x2,bins=30);
plt.legend(["first","second"],framealpha=3,frameon=False);#for frame for legend
```
