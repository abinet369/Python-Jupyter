```python
#1
n2=np.genfromtxt(r'C:\Users\abine\OneDrive\Documents\Datasets\Iris\iras.csv',delimiter=',',dtype='float',usecols=[0,1,2,3])
n2
#print(n2[0:,[0,2]])
n3=(n2[0:,[0,2]])
print(n3)
n3[np.isnan(n3)]=0
me=np.mean(n3[0:,[0]])
print("sepal length mean =",me)
me1=np.mean(n3[0:,[1]])   # here we have to give the 1 instead of the index number
print("petal length mean = " ,me1)
md=np.median(n3[0:,[0]])
print("sepal length median = ",md)
md1=np.median(n3[0:,[1]])
print("petal length median = ",md1)
sd=np.std(n3[0:,[0]])
print("sepal length standard deviation =",sd)
sd1=np.std(n3[0:,[1]])
print("petal length standard deviation =",sd1)
```
